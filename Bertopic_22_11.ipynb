```python
#импортирование библиотек
import pandas as pd
import numpy as np
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN
from sentence_transformers import SentenceTransformer
```

```python
#загрузка датесета
df = pd.read_excel('/content/Лист Microsoft Excel.xlsx')
docs = df['19 СОБЫТИЯ'].tolist()
```

```python
#модель для эмбендингов
sentence_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
```

```python
#настройки
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
#прописываем UMAP
umap_model = UMAP(
    n_neighbors=15,           # Уменьшено для лучшего локального разделения
    n_components=30,         # Сохраняет больше семантики
    min_dist=0.3,           # Более равномерное распределение
    metric='cosine',
    random_state=42
)
#прописываем HDBSCAN
hdbscan_model = HDBSCAN(
    min_cluster_size=5,     # Средний размер для стабильных кластеров
    min_samples=5,          # Баланс между шумом и кластерами
    #cluster_selection_epsilon=0.1,  # Объединяет близкие группы
    metric='euclidean',
    prediction_data=True
)
#сама модель
topic_model_21 = BERTopic(
    embedding_model=sentence_model, language="russian",
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    min_topic_size=8,       # Разрешает более мелкие, но значимые темы
    n_gram_range=(1, 2),
    top_n_words=10,
    calculate_probabilities=True
)
```

```python
topics21, probs21 = topic_model_21.fit_transform(docs)
topic_model_21.get_topic_info()
```

```python
df_after_analiz21 = pd.DataFrame({"Document": docs, "Topic": topics21}) #создаем датафрейм для просмотра кластеров
```

Варианты нахождения смешанных тем: 1) выведение документов с самой низкой максимальной вероятностью отнесения к теме, 2)

```python
# Создаем DataFrame из массива вероятностей
df_probabilities = pd.DataFrame(probs21)

# Добавляем исходные документы и итоговую тему в DataFrame для контекста
df_probabilities['Document'] = docs
df_probabilities['Assigned_Topic_ID'] = topics21

# Находим максимальную вероятность для каждого документа
# Выбираем только столбцы с вероятностями тем (все, кроме последних двух столбцов с метаданными)
prob_columns = df_probabilities.columns[:-2]
df_probabilities['Max_Probability'] = df_probabilities[prob_columns].max(axis=1)

# Сортировка по возрастанию максимальной вероятности (самые "неуверенные" сверху)
df_ambiguous_docs = df_probabilities.sort_values(by='Max_Probability', ascending=True)

# Настройка отображения Pandas для читаемости
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.5f' % x)

print("\n--- Топ-20 самых неоднозначных документов (с наименьшей максимальной вероятностью): ---")
# Выводим только нужные для анализа столбцы
print(df_ambiguous_docs[['Document', 'Assigned_Topic_ID', 'Max_Probability']].head(20))
```

```python
# 1. Находим ТОП-2 вероятности для каждой строки
# argsort сортирует индексы, берем последние 2 (самые большие)
top_two_indices = np.argsort(-df_probabilities[prob_columns].values, axis=1)[:, :2]

# 2. Извлекаем значения этих вероятностей
max_probs = df_probabilities[prob_columns].values[np.arange(len(df_probabilities)), top_two_indices[:, 0]]
second_max_probs = df_probabilities[prob_columns].values[np.arange(len(df_probabilities)), top_two_indices[:, 1]]

# 3. Рассчитываем разницу между первой и второй вероятностью
df_probabilities['Prob_Difference'] = max_probs - second_max_probs

# 4. Сортируем по возрастанию разницы (чем меньше разница, тем ближе 2 темы друг к другу)
df_mixed_topics = df_probabilities.sort_values(by='Prob_Difference', ascending=True)

# 5. Выводим результаты
print("\n--- Топ-20 самых 'смешанных' документов (с близкими 1-й и 2-й вероятностями): ---")
# Также покажем две основные темы
df_mixed_topics['Top_Topic_ID'] = top_two_indices[:, 0]
df_mixed_topics['Second_Topic_ID'] = top_two_indices[:, 1]


print(df_mixed_topics[['Document', 'Assigned_Topic_ID', 'Top_Topic_ID', 'Second_Topic_ID', 'Prob_Difference']].head(20))
```

```python
# Устанавливаем порог разницы вероятностей (например, 0.05, 0.1 или 0.15)
# Это число определяет, насколько близко должна быть вторая тема к первой, чтобы документ считался "смешанным"
threshold = 0.001

# Фильтруем DataFrame, оставляя только те документы, где разница между топ-1 и топ-2 вероятностями мала
mixed_docs_filtered = df_mixed_topics[df_mixed_topics['Prob_Difference'] < threshold].copy()

# Дополнительно: вы можете отфильтровать документы, которые были отнесены к кластеру -1,
# если хотите видеть только те, что попали в основные кластеры:
# mixed_docs_filtered = mixed_docs_filtered[mixed_docs_filtered['Assigned_Topic_ID'] != -1]


print(f"Найдено {len(mixed_docs_filtered)} документов, где разница вероятностей меньше {threshold}.")

print(mixed_docs_filtered[['Document', 'Assigned_Topic_ID', 'Top_Topic_ID', 'Second_Topic_ID', 'Prob_Difference']].head(20))
```

```python
import pandas as pd
from collections import Counter

# Функция для получения топ-слов темы
def get_topic_words(topic_model, topic_id):
    try:
        topic_words = topic_model.get_topic(topic_id)
        if topic_words:
            return ', '.join([word for word, score in topic_words[:5]])
        else:
            return 'нет данных'
    except:
        return 'ошибка'

# Функция для получения размера кластера
def get_cluster_size(topic_id, all_topics):
    return Counter(all_topics).get(topic_id, 0)

# Функция для безопасного преобразования кодов (обрабатывает и целые и float)
def safe_code_convert(code_str):
    try:
        # Пробуем преобразовать в float, затем в int (для случаев типа "70.0")
        return int(float(code_str))
    except (ValueError, TypeError):
        return code_str  # Если не получается, оставляем как есть

# Функция проверки совпадения с эталоном
def check_match(gt_codes, predicted_code):
    if not gt_codes:
        return 'нет эталона'

    # Преобразуем predicted_code к тому же формату, что и gt_codes
    predicted_code_num = safe_code_convert(predicted_code)

    # Преобразуем все gt_codes к числам для сравнения
    gt_codes_nums = [safe_code_convert(code) for code in gt_codes]

    return 'ДА' if predicted_code_num in gt_codes_nums else 'НЕТ'

# Функция для расчета точности внутри кластера
def calculate_cluster_precision(topic_docs):
    correct = 0
    total = 0

    for _, row in topic_docs.iterrows():
        if row['Эталонные_коды'] != 'нет':
            total += 1
            # Безопасно преобразуем коды
            gt_codes = [safe_code_convert(x.strip()) for x in row['Эталонные_коды'].split(',')]
            predicted_code_num = safe_code_convert(row['Предсказанный_код'])

            if predicted_code_num in gt_codes:
                correct += 1

    return round(correct / total, 3) if total > 0 else 0

# Создаем DataFrame с полной информацией о кластеризации
clustering_results = []

for i, (topic, doc_text) in enumerate(zip(topics21, docs)):
    predicted_code = manual_single_code_mapping.get(topic, 510)

    # Получаем эталонные коды для этого документа
    gt_codes = ground_truth[i] if i < len(ground_truth) else []

    clustering_results.append({
        '№ документа': i + 1,
        'Текст документа': doc_text,
        'Тема_BERTopic': topic,
        'Предсказанный_код': predicted_code,
        'Эталонные_коды': ', '.join(map(str, gt_codes)) if gt_codes else 'нет',
        'Топ_слова_темы': get_topic_words(topic_model_21, topic),
        'Размер_кластера': get_cluster_size(topic, topics21),
        'Совпадение_с_эталоном': check_match(gt_codes, predicted_code)
    })

# Создаем основной DataFrame
df_clusters = pd.DataFrame(clustering_results)

# Создаем сводную таблицу по кластерам
cluster_summary = []
for topic_id in sorted(set(topics21)):
    topic_docs = df_clusters[df_clusters['Тема_BERTopic'] == topic_id]
    size = len(topic_docs)

    # Самые частые эталонные коды в кластере (с безопасным преобразованием)
    all_gt_codes = []
    for codes in topic_docs['Эталонные_коды']:
        if codes != 'нет':
            # Безопасно преобразуем все коды
            converted_codes = [safe_code_convert(x.strip()) for x in codes.split(',')]
            all_gt_codes.extend(converted_codes)

    gt_code_counts = Counter(all_gt_codes)
    most_common_gt = gt_code_counts.most_common(3) if gt_code_counts else []

    cluster_summary.append({
        'Тема_BERTopic': topic_id,
        'Размер_кластера': size,
        'Предсказанный_код': manual_single_code_mapping.get(topic_id, 510),
        'Топ_эталонные_коды': str(most_common_gt),
        'Топ_слова': get_topic_words(topic_model_21, topic_id),
        'Точность_в_кластере': calculate_cluster_precision(topic_docs)
    })

df_summary = pd.DataFrame(cluster_summary)

# Создаем Excel файл с несколькими листами
with pd.ExcelWriter('bertopic_clustering_results.xlsx', engine='openpyxl') as writer:
    # Лист 1: Подробные результаты по документам
    df_clusters.to_excel(writer, sheet_name='Все_документы', index=False)

    # Лист 2: Сводка по кластерам
    df_summary.to_excel(writer, sheet_name='Сводка_по_кластерам', index=False)

    # Лист 3: Сопоставление тем и кодов
    mapping_df = pd.DataFrame([
        {'Тема_BERTopic': topic, 'Предсказанный_код': code}
        for topic, code in manual_single_code_mapping.items()
    ])
    mapping_df.to_excel(writer, sheet_name='Сопоставление_тем_кодов', index=False)

    # Лист 4: Статистика качества
    quality_stats = pd.DataFrame([{
        'Всего_документов': len(docs),
        'Всего_кластеров': len(set(topics21)),
        'Доля_шума_%': round((Counter(topics21).get(-1, 0) / len(docs)) * 100, 2),
        'Средний_размер_кластера': round(df_summary['Размер_кластера'].mean(), 1),
        'Минимальный_размер': df_summary['Размер_кластера'].min(),
        'Максимальный_размер': df_summary['Размер_кластера'].max(),
        'Микро_Precision': round(micro_precision_single, 3),
        'Микро_Recall': round(micro_recall_single, 3),
        'Микро_F1': round(micro_f1_single, 3)
    }])
    quality_stats.to_excel(writer, sheet_name='Статистика_качества', index=False)

print("✅ Excel файл 'bertopic_clustering_results.xlsx' успешно создан!")
print(f"📊 Содержит {len(df_clusters)} документов в {len(df_summary)} кластерах")

# Дополнительно: создаем файл с примерами документов из каждого кластера
def create_examples_file():
    examples = []

    for topic_id in sorted(set(topics21)):
        topic_docs = df_clusters[df_clusters['Тема_BERTopic'] == topic_id]

        if len(topic_docs) > 0:
            # Берем первые 3 документа из кластера как примеры
            sample_docs = topic_docs.head(3)

            for idx, row in sample_docs.iterrows():
                examples.append({
                    'Тема_BERTopic': topic_id,
                    'Размер_кластера': len(topic_docs),
                    'Предсказанный_код': row['Предсказанный_код'],
                    'Пример_документа': row['Текст документа'][:200] + '...' if len(row['Текст документа']) > 200 else row['Текст документа'],
                    'Эталонные_коды': row['Эталонные_коды'],
                    'Совпадение': row['Совпадение_с_эталоном']
                })

    df_examples = pd.DataFrame(examples)
    df_examples.to_excel('bertopic_cluster_examples.xlsx', index=False)
    print("✅ Дополнительный файл 'bertopic_cluster_examples.xlsx' создан!")

create_examples_file()

# Показываем статистику
print(f"\n📈 СТАТИСТИКА ВЫГРУЗКИ:")
print(f"   - Всего документов: {len(df_clusters)}")
print(f"   - Всего кластеров: {len(set(topics21))}")
print(f"   - Размер самого большого кластера: {df_summary['Размер_кластера'].max()}")
print(f"   - Размер самого маленького кластера: {df_summary['Размер_кластера'].min()}")
print(f"   - Доля шума: {Counter(topics21).get(-1, 0) / len(docs) * 100:.1f}%")

# Дополнительно: создаем CSV для быстрого просмотра
df_clusters.to_csv('bertopic_clusters_quick_view.csv', index=False, encoding='utf-8-sig')
print("✅ CSV файл 'bertopic_clusters_quick_view.csv' создан для быстрого просмотра!")

# Выводим топ-5 самых больших кластеров для быстрого обзора
print(f"\n🏆 ТОП-5 САМЫХ БОЛЬШИХ КЛАСТЕРОВ:")
top_clusters = df_summary.nlargest(5, 'Размер_кластера')
for _, row in top_clusters.iterrows():
    print(f"   Тема {row['Тема_BERTopic']}: {row['Размер_кластера']} док. → код {row['Предсказанный_код']} ({row['Топ_слова'][:50]}...)")
```
