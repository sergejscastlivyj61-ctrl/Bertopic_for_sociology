```python
#импортирование библиотек
import pandas as pd
import numpy as np
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN
from sentence_transformers import SentenceTransformer
```

```python
#загрузка датесета
df = pd.read_excel('/content/Лист Microsoft Excel.xlsx')
docs = df['19 СОБЫТИЯ'].tolist()
```

```python
#модель для эмбендингов
sentence_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
```

```python
#настройки
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
# УРОВЕНЬ 1: Крупные кластеры
umap_large = UMAP(
    n_neighbors=10,
    n_components=20,           # Увеличиваем размерность для сохранения информации
    min_dist=0.4,             # Более компактные кластеры
    metric='cosine',
    random_state=42
)

hdbscan_large = HDBSCAN(
    min_cluster_size=20,       # Крупные кластеры
    min_samples=10,
    metric='euclidean',
    cluster_selection_epsilon=0.2,  # Объединяет близкие кластеры
    prediction_data=True
)

topic_model_large = BERTopic(
    embedding_model=sentence_model,
    umap_model=umap_large,
    hdbscan_model=hdbscan_large,
    min_topic_size=20,         # Соответствует min_cluster_size
    language="russian",
    n_gram_range=(1, 2),
    top_n_words=8,
    calculate_probabilities=True
)
```

```python
topics_large, probs_large = topic_model_large.fit_transform(docs)
topic_model_large.get_topic_info()
```

```python
df_after_analiz_large = pd.DataFrame({"Document": docs, "Topic": topics_large}) #создаем датафрейм для просмотра кластеров
```

```python
topic_large = 1 #id класса
documents_in_topic_large = df_after_analiz_large[df_after_analiz_large['Topic'] == topic_large]['Document'].tolist()
documents_in_topic_large
```

```python
# Копируем исходные темы
final_topics = topics_large.copy()

# Индексы документов тем -1 и 0
problem_indices = [i for i, t in enumerate(topics_large) if t in [-1, 0]]
problem_docs = [docs[i] for i in problem_indices]

if len(problem_docs) > 0:
    # Создаем отдельную BERTopic модель для повторной кластеризации проблемных тем
    umap_detail = UMAP(
        n_neighbors=10,
        n_components=15,
        min_dist=0.1,
        metric='cosine',
        random_state=42
    )

    hdbscan_detail = HDBSCAN(
        min_cluster_size=15,
        min_samples=15,
        metric='euclidean',
        cluster_selection_method='eom',
        prediction_data=True
    )

    topic_model_detail = BERTopic(
        embedding_model=sentence_model,
        umap_model=umap_detail,
        hdbscan_model=hdbscan_detail,
        min_topic_size=15,
        language="russian",
        n_gram_range=(1, 2),
        top_n_words=12,
        calculate_probabilities=True
    )

    # Обучаем модель на проблемных документах
    detail_topics, detail_probs = topic_model_detail.fit_transform(problem_docs)
```

```python
topic_model_detail.get_topic_info()
```

```python
df_after_analiz_detail = pd.DataFrame({"Document": problem_docs, "Topic": detail_topics}) #создаем датафрейм для просмотра кластеров
```

```python
topic_detail = -1 #id класса
documents_in_topic_detail = df_after_analiz_detail[df_after_analiz_detail['Topic'] == topic_detail]['Document'].tolist()
documents_in_topic_detail
```
